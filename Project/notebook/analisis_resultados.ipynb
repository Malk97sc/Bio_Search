# Análisis de Resultados – Metaheurísticas para el Problema de la Mochila

Este notebook presenta un análisis exploratorio de los resultados obtenidos al aplicar dos algoritmos bioinspirados:
- **Simulated Annealing (SA)**
- **Ant Colony Optimization (ACO)**

Se incluyen:
- Estadísticas por configuración
- Visualización de convergencia y desempeño
- Comparaciones individuales y globales

```python
import os
import pandas as pd
from IPython.display import display, Image
```

## Resultados por configuración

```python
base_path = "../results"
config_dirs = ["original_SA_vs_ACO", "competitiva_SA_vs_ACO", "exploratoria_SA_vs_ACO", "debil_SA_vs_ACO"]

for config in config_dirs:
    print(f"\nResultados para configuracion: {config} ")
    nombre = config.split("_SA_vs_ACO")[0]
    ruta = os.path.join(base_path, config)
    
    resumen_path = os.path.join(ruta, f"resumen_estadistico_{nombre}.csv")
    mejor_sol_path = os.path.join(ruta, f"mejor_solucion_{nombre}.csv")

    if os.path.exists(resumen_path):
        print("\nResumen estadistico:")
        display(pd.read_csv(resumen_path))

    if os.path.exists(mejor_sol_path):
        print("\nMejor solucion encontrada:")
        display(pd.read_csv(mejor_sol_path))

    for grafico in ["convergencia.png", "mejores_valores.png", "histograma.png", "iters_convergencia.png"]:
        img_path = os.path.join(ruta, grafico)
        if os.path.exists(img_path):
            print(f"\n{grafico}")
            display(Image(filename=img_path))
```

## Análisis de ejecuciones individuales

```python
for metodo in ["sa_original", "aco_competitiva"]:
    ruta_indiv = os.path.join(base_path, "test_individual", metodo)
    print(f"\n{metodo.upper()}")
    csv_path = os.path.join(ruta_indiv, "resumen_individual.csv")
    if os.path.exists(csv_path):
        display(pd.read_csv(csv_path))
    img_path = os.path.join(ruta_indiv, "convergencia_individual.png")
    if os.path.exists(img_path):
        display(Image(filename=img_path))
```
